```python
from ultralytics import YOLO

# Load a pretrained YOLOv8n model
model = YOLO("best.pt")

# Run inference on 'bus.jpg' with arguments
results = model.predict("./testimages", save=True, conf=0.25)
names = model.names
```

```python
i = 0
map = {}
for r in results:
    for c in r.boxes.cls:
        if names[int(c)] not in map:
            map[names[int(c)]] = 1
        else:
            map[names[int(c)]] += 1
    i += 1
print(map)
```
